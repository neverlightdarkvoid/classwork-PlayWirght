```python
#list的typehint
#list
rabbits:list[str] = ["Flopsy","Mopsy","Cottontail","Peter"]

for rabbit in rabbits:
    print(rabbit)
```

```python
#dict的typehint

country_code:dict[str,str] = {"tw":"Taiwan","jp":"Japan","hk":"Hong Kong","kr":"Korea"}
for code in country_code:
    print(f"key:{code}")
    print(f"value:{country_code[code]}")
    print("================================")
```
