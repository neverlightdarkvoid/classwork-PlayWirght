```python
#listзҡ„typehint
#list
rabbits:list[str] = ["Flopsy","Mopsy","Cottontail","Peter"]

for rabbit in rabbits:
    print(rabbit)
```

```python
#dictзҡ„typehint

country_code:dict[str,str] = {"tw":"Taiwan","jp":"Japan","hk":"Hong Kong","kr":"Korea"}
for code in country_code:
    print(f"key:{code}")
    print(f"value:{country_code[code]}")
    print("================================")
```

```python
#ијёе…ҘиӢұж–Үеӯ—жҜҚ (жҢү Q йӣўй–Ӣ)
while True:
    stuff = input("и«Ӣијёе…Ҙе°ҸеҜ«иӢұж–Үеӯ—жҜҚ[жҢүqжңғйӣўй–Ӣ]:")
    if stuff == 'q':
        break
    print(stuff.capitalize())
print("зЁӢејҸе·Ізөҗжқҹ")
```

```python
#зҙҜеҠ жӯЈеҒ¶ж•ё
#иЁӯиЁҲдёҖеҖӢзЁӢејҸи®“дҪҝз”ЁиҖ…ијёе…Ҙж•ёеҖјпјҢеҸӘжңүжӯЈеҒ¶ж•ёеҖјжңғиў«еҠ зёҪпјҢиӢҘијёе…ҘеҘҮж•ёеүҮи·ійҒҺпјҢијёе…ҘиІ ж•ёеүҮзөҗжқҹзЁӢејҸпјҡ
total:int = 0
num:int = 0
while True:
    input_value = int(input(f"и«Ӣијёе…Ҙз¬¬{num + 1}еҖӢж•ёеҖј:"))    
    if input_value < 0:
        break
        
    num += 1

    if input_value % 2 == 1:
        continue

    total += input_value

print(f"ијёе…Ҙзҡ„ж¬Ўж•ёжҳҜ{num}, пјҢжүҖжңүијёе…Ҙзҡ„жӯЈеҒ¶ж•ёзҡ„еҠ зёҪжҳҜ:{total}")
```
