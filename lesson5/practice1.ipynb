```python
#巢狀判斷
furry = True
small = True

if furry:
    print("這是有毛的")
    if small:
        print("這是小的")
        print("它是貓")
    else:
        print("這是大的")
        print("它是熊")
else:
    print("這是代表沒有毛")
    if small:
        print("這是小的")
        print("它是小蜥蜴")
    else:
        print("這是大的")
        print("它是人類或沒毛的熊")
```

```python
# 如果 x 不是負數，則傳回值為 x ** (1 / y)
# 如果 x 是負數而且 y 為偶數，則傳回值為 "虛數"
# 如果 x 是負數而且 y 為奇數，則傳回值為 -(-x) ** (1 / y)

x = int(input("請輸入 x:"))
y = int(input("請輸入 y:"))

if x >= 0:
    root = x ** (1/y)
else:
    if y % 2 == 0:
        root = "虛數"
    else:
        root = -(-x) ** (1 / y)

print("root=", root)
```
