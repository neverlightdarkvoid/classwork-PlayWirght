```python
# 如果 x 不是負數，則傳回值為 x ** (1 / y)
# 如果 x 是負數而且 y 為偶數，則傳回值為 "虛數"
# 如果 x 是負數而且 y 為奇數，則傳回值為 -(-x) ** (1 / y)

x = int(input("請輸入 x:"))
y = int(input("請輸入 y:"))

if x >= 0:
    root = x ** (1/y)
else:
    if y % 2 == 0:
        root = "虛數"
    else:
        root = -(-x) ** (1 / y)

print("root=", root)
```

### 題目敘述
百貨公司在做週年慶折扣。請寫一支程式,先請使用者輸入「消費金額」,再詢問使用者「是否為會員?(y,n)」。

如果是會員(輸入 y),全部金額打 9 折(乘以 0.9)。但公司規定:折扣後的最低應付金額不得低於 100 元,如果打折後低於 100,就以 100 元計算。

最後輸出使用者實際要付的金額(四捨五入到整數)。

```python
# 百貨公司週年慶:會員打 9 折
# 折扣後最低不得低於 100 元

money = int(input("請輸入消費金額:"))
is_member = input("是否為會員?(y,n)")
is_discount = False

# 請完成:如果是會員(is_member 等於 "y"),才進入打折流程
if is_member == 'y':
    # 請完成:把 money 打 9 折(乘以 0.9)
    money_ori = money
    money = money * 0.9
    is_discount = True
    
    # 請完成:如果打折後金額小於 100,就把 money 設成 100
    if is_discount == True and money <= 100:
        money = 100
        if money_ori <= money:
            money = money_ori

# 請完成:用 round() 把金額四捨五入後印出
print(round(money))
```
