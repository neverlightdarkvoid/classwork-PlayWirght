### 題目敘述
便利商店推出「飲料杯數折扣」活動,請寫一支程式,讓使用者輸入一次購買的飲料杯數,程式依照下列規則判斷並輸出折扣方案:  
  
20 杯(含)以上:打 7 折  
10 杯(含)~ 19 杯:打 8 折  
5 杯(含)~ 9 杯:打 9 折  
1 杯(含)~ 4 杯:原價(不打折)  
0 杯:輸出「沒有購買任何飲料」  

```python
cups = int(input("請輸入購買的飲料杯數："))

if cups >= 20:
    discount = '打 7 折'
elif cups >= 10:
    discount = '打 8 折'
elif cups >= 5:
    discount = '打 9 折'
elif cups >= 1:
    discount = '原價'
else:
    discount = '沒有購買任何飲料'


print("折扣方案：", discount)
```
